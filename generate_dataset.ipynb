```python
import pandas as pd

# Caminho para o arquivo Excel
file_path = 'Matriz_de_Insumo_Produto_2015_Nivel_12.xls'

# Ler todas as planilhas do arquivo Excel
xls = pd.ExcelFile(file_path)

# Obter os nomes das planilhas
sheet_names = xls.sheet_names

sheet_names
```

```python
def clean_sheet(sheet_name):
    df = pd.read_excel(file_path, sheet_name=sheet_name, header=None)

    new_sheet_name = df.iloc[0, 0]

    # Encontrar a primeira ocorrência de "Fonte" e truncar o DataFrame
    fonte_index = df[df.iloc[:, 0].astype(str).str.startswith('Fonte')].index
    fonte_index

    if not fonte_index.empty:
        df = df[:fonte_index[0]]
    
    # Remover colunas e linhas vazias
    df = df.dropna(axis=1, how='all')
    df = df.dropna(how='all')


    df = df.ffill(axis=0) 
    df.reset_index(drop=True, inplace=True)

    df.columns = [title.replace('\n', ' ') for title in df.loc[2]]
    df = df.iloc[3:]
    df.reset_index(drop=True, inplace=True)

    
    return new_sheet_name, df

dfs = {}

# Processar as primeiras 10 abas da planilha (tabelas base)
for sheet in sheet_names[:10]:
    new_name, cleaned_df = clean_sheet(sheet)
    dfs[new_name] = cleaned_df
    
# Exibir os nomes das novas planilhas e os primeiros 5 registros de cada DataFrame limpo
for sheet in list(dfs.keys())[:10]:
    print(f'Planilha: {sheet}')
    print(dfs[sheet].head())
    print()
```
